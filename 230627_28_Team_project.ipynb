### 공간 시각화에서 notnull 메서드를 통해 'NaN' 값들은 전부 제외한 상태로 처리

```python
# 기본 지도 시각화 1

import folium


# 지도 초기화
m = folium.Map(location=[37.5502, 126.982], zoom_start=8)  # 초기 위치와 줌 레벨 설정
                                                            # 초기 위치 -> 서울의 중심 location=[37.5502, 126.982]
# 데이터프레임 순회하며 위치 표시
for index, row in df.iterrows():
    if pd.notnull(row['위도']) and pd.notnull(row['경도']): # 각 행의 위도와 경도 값이 유효한 경우에만 위치를 표시합니다. (pd.notnull() 함수를 사용하여 유효성을 확인합니다.)
        folium.Marker(location=[row['위도'], row['경도']]).add_to(m) # 위치를 표시하기 위해 folium.Marker를 생성하고 지도에 추가합니다.

# 지도 출력
m

# 즉, 위 코드는 데이터프레임의 각 행에서 유효한 위도와 경도 값을 가져와 해당 위치에 마커를 표시하여 지도를 생성합니다. 유효하지 않은 값을 가진 행은 위치 표시를 하지 않습니다.
```

```python
# 기본 지도 시각화 2 서울-경기 부근 지도 제한

import pandas as pd
import folium


# 서울-경기도 부근 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=10)

# 데이터프레임 순회하며 위치 표시
for index, row in df.iterrows():
    if pd.notnull(row['위도']) and pd.notnull(row['경도']):
        lat, lon = row['위도'], row['경도']
        # 서울-경기도 부근인 경우에만 표시
        if 37.0 <= lat <= 38.5 and 126.0 <= lon <= 127.5:
            folium.Marker(location=[lat, lon]).add_to(m)

# 지도 출력
m
```

```python
# 클러스터 마커 시각화1
from folium.plugins import MarkerCluster


m = folium.Map(location=[37.5502, 126.982], zoom_start=10)  # 초기 위치와 줌 레벨 설정

# 클러스터맵 생성
marker_cluster = MarkerCluster().add_to(m)

# 데이터프레임 순회하며 위치 표시
for index, row in df.iterrows():
    if pd.notnull(row['위도']) and pd.notnull(row['경도']):
        folium.Marker(location=[row['위도'], row['경도']]).add_to(marker_cluster)

# 지도 출력
m
```

```python
# 클러스터 마커 시각화2 서울-경기 부근 지도 제한
from folium.plugins import MarkerCluster

# 서울-경기도 부근 좌표 범위 설정
seoul_lat, seoul_lon = 37.5665, 126.9780
bbox = [
    [seoul_lat - 0.5, seoul_lon - 0.5],
    [seoul_lat + 0.5, seoul_lon + 0.5]
]

# 지도 초기화
m = folium.Map(location=[seoul_lat, seoul_lon], zoom_start=10)  # 초기 위치와 줌 레벨 설정

# 클러스터맵 생성
marker_cluster = MarkerCluster().add_to(m)

# 데이터프레임 순회하며 위치 표시
for index, row in df.iterrows():
    if pd.notnull(row['위도']) and pd.notnull(row['경도']): # 각 행의 위도와 경도 값이 유효하고, 서울-경기도 부근의 좌표 범위에 해당하는 경우에만 마커를 표시합니다.
        lat, lon = row['위도'], row['경도']
        if bbox[0][0] <= lat <= bbox[1][0] and bbox[0][1] <= lon <= bbox[1][1]:
            folium.Marker(  # folium.Marker를 사용하여 해당 위치에 마커를 생성하고, 클러스터맵에 추가합니다.
                location=[lat, lon],
                icon=folium.Icon(color='blue', icon='circle'),
                popup=row['병원명/센터명']   # 마커의 아이콘은 파란색 원으로 설정되며, 팝업에는 해당 의료기관의 이름을 표시합니다.
            ).add_to(marker_cluster)

# 지도 출력
m
```
